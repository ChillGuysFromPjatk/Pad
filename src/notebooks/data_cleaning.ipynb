```python
import pandas as pd
import re
import numpy as np
```

```python
input_csv = 'raw/output_ebayspidercat_2.csv'
```

```python
df = pd.read_csv(input_csv)
df
```

```python
df.dropna(inplace=True)
df
```

```python
#usuwanie przedrostka
df['country'] = df['country'].str[3:]
df
```

```python
#usuwanie podwojonych cen (biore tylko pierwsza)
def extract_first_price(price):
    match = re.match(r'^\d{1,3}(,\d{2})? zł', price)
    return match.group(0) if match else price

df['price'] = df['price'].apply(lambda x: re.search(r'\d{1,6},\d{2} zł', x).group(0))
df
```

```python
#zamiana ceny na wartość liczbową
df['price_pln'] = df['price'].str.replace(',', '.').str.replace(' zł', '').astype(float)
df.drop(columns=['price'], inplace=True)
df
```

```python
df.dtypes
```

```python
#zamiana daty na datetime
month_map = {
    'sty': '01', 'lut': '02', 'mar': '03', 'kwi': '04', 'maj': '05', 'cze': '06',
    'lip': '07', 'sie': '08', 'wrz': '09', 'paź': '10', 'lis': '11', 'gru': '12'
}
df['sale_date'] = df['sale_date'].str.replace('Sprzedane  ', '', regex=False)
for pol, num in month_map.items():
    df['sale_date'] = df['sale_date'].str.replace(f' {pol} ', f'-{num}-', regex=False) #.str.replace(' ', '')
df
```

```python
df['sale_date'] = pd.to_datetime(df['sale_date'], format='%d-%m-%Y')
df
```

```python
#wyodrebnienie producenta
df['producer'] = df['model'].str.split(' ').str[0].str.replace('New', 'Nintendo')
df['model'] = df['model'].str.split(' ', n=1).str[1].str.replace('Nintendo', 'New')
df.sample(20)
```

```python
df['model'].unique()
```

```python
#grupowanie roznych rodzajow tej samej konsoli
model_group_map = {
    'Xbox 360': ['Xbox 360', 'Xbox 360 Arcade', 'Xbox 360 Core', 'Xbox 360 E', 'Xbox 360 Elite', 'Xbox 360 Premium', 'Xbox 360 Pro', 'Xbox 360 S'],
    'Xbox One': ['Xbox One', 'Xbox One S', 'Xbox One X'],
    'Xbox Series': ['Xbox Series S', 'Xbox Series X'],
    'NES': ['NES', 'NES Classic Edition'],
    'SNES': ['SNES', ' Super NES Classic Edition'],
    'Wii': ['Wii', 'Wii mini'],
    'Wii U': ['Wii U - Basic', 'Wii U - Deluxe'],
    'Switch': ['Switch', 'Switch (OLED Model)', 'Switch Lite'],
    'Game Boy': ['Game Boy Color', 'Game Boy Light', 'Game Boy Micro', 'Game Boy Pocket'],
    'Game Boy Advance': ['Game Boy Advance', 'Game Boy Advance SP'],
    'DS': ['DS', 'DSi', 'DSi XL', 'DS Lite'],
    '2DS': ['2DS', '2DS LL', 'New 2DS XL'],
    '3DS': ['3DS', '3DS XL', 'New 3DS', 'New 3DS LL', 'New 3DS XL'],
    'PlayStation 1': ['PlayStation 1', 'PlayStation Classic'],
    'PlayStation 2': ['PlayStation 2', 'PlayStation 2 - Slim'],
    'PlayStation 3': ['PlayStation 3', 'PlayStation 3 - Slim', 'PlayStation 3 - Super Slim'],
    'PlayStation 4': ['PlayStation 4', 'PlayStation 4 Pro', 'PlayStation 4 Slim'],
    'PlayStation 5': [ 'PlayStation 5 Blu-Ray Edition', 'PlayStation 5 Digital Edition', 'PlayStation 5 Slim Blu-Ray Edition', 'PlayStation 5 Slim Digital Edition'],
    'PSP': [ 'PlayStation Portable', 'PSP-1000', 'PSP-1001', 'PSP-2000', 'PSP-3000', 'PSP-3001', 'PSP-E1000'],
    'PS Vita': ['PS Vita - PCH-1000', 'PS Vita - PCH-1001', 'PS Vita - PCH-1101', 'PS Vita - PCH-2000', 'PS Vita PCH-2001']
}

model_to_group = {model: group for group, models in model_group_map.items() for model in models}
df['model_group'] = df['model'].map(model_to_group).fillna(df['model'])

#df_x360 = df[df['model_group'] == 'Xbox 360']
#df_x360
df.sample(20)
```

```python
#dodanie info czy handheld czy nie
handhelds = {'Switch', 'Game Boy', 'Game Boy Advance', 'DS', '2DS,', '3DS', 'PSP', 'PS Vita', 'Steam Deck'}
home_consoles = {'Xbox', 'Xbox 360', 'Xbox One', 'Xbox Series', 'NES', 'SNES', '64', 'GameCube', 'Wii', 'Wii U', 'Switch', 'PlayStation 1', 'PlayStation 2', 'PlayStation 3', 'PlayStation 4', 'PlayStation 5', 'Steam Deck'}

df['handheld'] = df['model_group'].apply(lambda x: 1 if x in handhelds else 0)
df['home_console'] = df['model_group'].apply(lambda x: 1 if x in home_consoles else 0)

df.sample(20)
```

```python
#przypisanie generacji konsol
generation_mapping = {
    'GameCube': 6,
    'Wii': 7,
    'Wii U': 8,
    'Switch': 8,
    'Xbox': 6,
    'Xbox 360': 7,
    'Xbox One': 8,
    'Xbox Series': 9,
    'PlayStation 1': 5,
    'PlayStation 2': 6,
    'PlayStation 3': 7,
    'PlayStation 4': 8,
    'PlayStation 5': 9,
    'PSP': 6,
    'PS Vita': 7,
    'Steam Deck': 9
}

df['generation'] = df['model_group'].map(generation_mapping).fillna(-1).astype(int)

df.sample(20)
```

```python
#seller info
df[['seller_name', 'seller_num_comments', 'seller_rating']] = df['seller'].str.split(' ', expand=True)
df['seller_num_comments'] = df['seller_num_comments'].str.replace(r'\D', '', regex=True).astype(int)
df['seller_rating'] = df['seller_rating'].str.replace('%', '').str.replace(',', '.').astype(float)
df.drop(columns = ['seller'], inplace=True)
df.sample(20)
```

```python
#wyodrebnienie regionow swiata [ogloszenia sa w ogromnej wiekszosci z US, GB i JP i do nich grupuje reszte [chyba sensownie]]
#df['country'].unique()


country_map = {
    'AM': ['Stany Zjednoczone', 'Kanada', 'Meksyk', 'Portoryko', 'Jamajka', 'Wenezuela', 'Brazylia', 'Chile'],
    'EU': ['Wielka Brytania', 'Belgia', 'Holandia', 'Litwa', 'Irlandia', 'Włochy', 'Dania', 'Grecja', 
           'Portugalia', 'Albania', 'Francja', 'Mołdawia', 'Ukraina', 'Norwegia', 'Hiszpania', 'Finlandia',  
           'Szwecja', 'Słowenia', 'Estonia', 'Bułgaria', 'Cypr', 'Malta', 'Rumunia', 'Bośnia i Hercegowina', 
           'Łotwa', 'Niemcy', 'Węgry', 'Słowacja', 'Maroko', 'Egipt', 'Algieria', 'Zjednoczone Emiraty Arabskie', 
           'Turcja', 'Kazachstan', 'Arabia Saudyjska', 'Bahrajn', 'Kuwejt', 'Jordania', 'Izrael'],
    'AS': ['Japonia', 'Chiny', 'Tajwan', 'Singapur', 'Korea Południowa', 'Malezja',  
           'Indonezja', 'Tajlandia', 'Wietnam', 'Sri Lanka', 'Pakistan', 
           'Indie', 'Hongkong', 'Filipiny', 'Australia'],
}

country_to_region = {country: region for region, countries in country_map.items() for country in countries}
df['region'] = df['country'].map(country_to_region)

df.sample(20)
```

```python
#dyskretyzacja stanu
df['condition'].unique()

condition_map = {
    'Fabrycznie nowy': 3,
    'Nowy (inny)': 3,
    'Odnowiony': 2,
    'Używany': 1,
    'Tylko na części': 0
}

df['condition_num'] = df['condition'].map(condition_map)

df.sample(20)
```

```python
#obliczanie seller score na podstawie oceny i komentarzy
#df['seller_num_comments'].describe()
#df['seller_rating'].describe()

df['seller_score'] = df['seller_rating'] * (1 + (np.log10(df['seller_num_comments'] + 1) / np.log10(df['seller_num_comments'].max() + 1)))
#normalizuje
df['seller_score'] = (df['seller_score'] - df['seller_score'].min()) / (df['seller_score'].max() - df['seller_score'].min())

df.sample(20)
```

```python
keywords = [
    'rare', 'limited', 'collect',
    'rzadk', 'limitowan', 'kolekcj'
]

keyword_pattern = r'|'.join([f'\\b{k}' for k in keywords])
df['collectible'] = df['title'].str.contains(keyword_pattern, case=False, na=False).astype(int)

df.sample(20)
```

```python
columns_to_drop_or_encode = ['model_group','condition','country','model','sale_date','title','producer','seller_name', 'region']

df_dropped_columns = df.drop(columns=columns_to_drop_or_encode)
df_dropped_columns.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df_encoded_columns = df.copy()
for column in columns_to_drop_or_encode:
    df_encoded_columns[column] = le.fit_transform(df_encoded_columns[column])
df_encoded_columns.head()
```
